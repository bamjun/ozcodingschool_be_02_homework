```python
from pymongo import MongoClient

# MongoDB 인스턴스에 연결
client = MongoClient('mongodb://localhost:27017/')
client
```

```python
db = client['public_data']
db
#use example_db
```

```python
# 콜렉션 선택 (없으면 자동 생성)
collection = db['users']
collection
```

```python
# 모든 문서 조회
for doc in collection.find():
    print(doc)

print('# 조건에 맞는 문서 조회 John0')
# 조건에 맞는 문서 조회
query = {"lang_code_id": "en"}
for doc in collection.find(query):
    print(doc)
```

```python
for doc in collection.find({}, {'new_address': 1, 'tag': 1}):
    print(doc)
```

```python
for doc in collection.find({'post_sn': {'$lte': 70}}):
    print(doc)
#post_sn 이 70보다 작은 수 조회
```

```python
for doc in collection.aggregate([{'$group': {'_id': '$lang_code_id', 'count': {'$sum': 1}}}]):
    print(doc)
```

```python
for doc in collection.find({'lang_code_id': 'ko'}):
    print(doc)
```

```python
db.users.insert_one({'new_address': '광화문', 'tag': '이순신동상'})
```

```python
for x in db.users.find({'new_address': '광화문'}):
    print(x)
```

```python
db.users.update_one({'new_address': '광화문'},{'$set': {'post_url': '여기로가면되나요?'}})
```

```python
for x in db.users.find({'new_address': '광화문'}):
    print(x)
```

```python
db.users.delete_one({'new_address': '광화문'})
```

```python
for x in db.users.find({'new_address': '광화문'}):
    print(x)
print('삭제완료')
```
