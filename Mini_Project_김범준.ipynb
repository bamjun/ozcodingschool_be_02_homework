```python
import requests
import random

# https://random-word-api.herokuapp.com/home
# API ì„¤ëª…


# API URL ì„¤ì • 3ê¸€ìì—ì„œ 6ê¸€ì ì‚¬ì´ëœë¤ ë‹¨ì–´
url = "https://random-word-api.herokuapp.com/word?length=" + str(random.randint(3, 6))

# API ìš”ì²­ì„ ë³´ë‚´ê³  ì‘ë‹µì„ ë°›ìŠµë‹ˆë‹¤.
response = requests.get(url)

# ì‘ë‹µ ë°ì´í„°ë¥¼ JSON í˜•íƒœë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
word = response.json()[0]

guessed = "_" * len(word)
word_completion = "_" * len(word)
guessed_letters = []
tries = 10

# ê²Œì„ ì§„í–‰
while tries > 0 and not word_completion == word:
    print(f"ë‚¨ì€ ì‹œë„: {tries}")
    print(f"ë‹¨ì–´: {word_completion}")
    print(f"ì´ë¯¸ ì‹œë„í•œ ê¸€ì: {', '.join(guessed_letters)}")

    # ì‚¬ìš©ìë¡œë¶€í„° ê¸€ì ì…ë ¥ ë°›ê¸°
    guess = input("ê¸€ìë¥¼ ì¶”ì¸¡í•´ë³´ì„¸ìš”: ").lower()

    # ê¸€ì ì¶”ì¸¡ ì²˜ë¦¬
    if len(guess) == 1 and guess.isalpha():
        if guess in guessed_letters:
            print("ì´ë¯¸ ì‹œë„í•œ ê¸€ìì…ë‹ˆë‹¤.")
        elif guess not in word:
            print("í‹€ë ¸ìŠµë‹ˆë‹¤.")
            guessed_letters.append(guess)
            tries -= 1
        else:
            print("ì •ë‹µì…ë‹ˆë‹¤!")
            guessed_letters.append(guess)
            word_as_list = list(word_completion)
            indices = [i for i, letter in enumerate(word) if letter == guess]
            for index in indices:
                word_as_list[index] = guess
            word_completion = "".join(word_as_list)
            if "_" not in word_completion:
                break
    else:
        print("ì˜¬ë°”ë¥¸ ê¸€ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”. í•œê¸€ìë§Œ ì…ë ¥í•˜ì„¸ìš”.")

# ê²Œì„ ê²°ê³¼
if word_completion == word:
    print(f"ì¶•í•˜í•©ë‹ˆë‹¤! ë‹¨ì–´ëŠ” '{word}'ì˜€ìŠµë‹ˆë‹¤!")
else:
    print(f"ê²Œì„ ì˜¤ë²„. ë‹¨ì–´ëŠ” '{word}'ì˜€ìŠµë‹ˆë‹¤.")
















# ê¸°ë³¸ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¼ íŒŒì´ì¬ ì„¤ì¹˜ì‹œ ìë™ìœ¼ë¡œ ì„¤ì¹˜ë©ë‹ˆë‹¤!
# random ëª¨ë“ˆì€ ëœë¤í•œ ìˆ«ìë¥¼ ìƒì„±í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.
# import random

# ê²Œì„ì— ì‚¬ìš©ë  ë‹¨ì–´ ëª©ë¡
words = ["apple", "banana", "orange", "grape", "lemon"]

# í–‰ë§¨ ê·¸ë¦¼
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   / \
     |
    ---""",
]


class HangmanGame:
    # í´ë˜ìŠ¤ê°€ ë§Œë“¤ì–´ ì§ˆ ë•Œ ìë™ìœ¼ë¡œ í˜¸ì¶œë˜ëŠ” ë©”ì†Œë“œ
    # (ìƒì„±ì ë©”ì†Œë“œë¼ê³  ë¶€ë¦…ë‹ˆë‹¤)
    def __init__(self):
        self.word = random.choice(words)  # ëœë¤ìœ¼ë¡œ ë‹¨ì–´ ì„ íƒ

        self.guesses = set()  # ì¶”ì¸¡ëœ ê¸€ì ì €ì¥

        self.attempts = 6  # ì‹œë„ íšŸìˆ˜

    def display(self):
        # ë‹¨ì–´ì˜ í˜„ì¬ ìƒíƒœ í‘œì‹œ
        result = ""

        # ìœ„ì—ì„œ ì„¤ì •í•œ ë‹¨ì–´ì˜ ê¸€ìë“¤ì„ í•˜ë‚˜ì”© ê°€ì ¸ì˜µë‹ˆë‹¤.
        for char in self.word:
            # ë§Œì•½ ì¶”ì¸¡ëœ ê¸€ìê°€ ì„¤ì •í•œ ë‹¨ì–´ì— í¬í•¨ë˜ì–´ ìˆë‹¤ë©´
            if char in self.guesses:
                # ì¶”ì¸¡ëœ ê¸€ìë¥¼ í‘œì‹œí•©ë‹ˆë‹¤.
                result += char
            else:
                # ì¶”ì¸¡ëœ ê¸€ìê°€ ì•„ë‹ˆë¼ë©´ _ë¥¼ í‘œì‹œí•©ë‹ˆë‹¤.
                result += "_"

        # ë‹¨ì–´ì˜ í˜„ì¬ ìƒíƒœë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
        return result

    def play(self):
        # ê²Œì„ì„ ì‹œì‘í•©ë‹ˆë‹¤.
        # ì‹œë„ íšŸìˆ˜ê°€ 0ì´ ë  ë•Œê¹Œì§€ ë°˜ë³µí•©ë‹ˆë‹¤.
        while self.attempts > 0:
            # í–‰ë§¨ ê·¸ë¦¼ì„ í‘œì‹œí•©ë‹ˆë‹¤.
            print(hangman_pics[6 - self.attempts])

            # í˜„ì¬ ë‹¨ì–´ì˜ ìƒíƒœë¥¼ í‘œì‹œí•©ë‹ˆë‹¤.
            # (ì¶”ì¸¡ëœ ê¸€ìëŠ” í‘œì‹œí•˜ê³ , ì¶”ì¸¡ë˜ì§€ ì•Šì€ ê¸€ìëŠ” _ë¡œ í‘œì‹œí•©ë‹ˆë‹¤.)
            # ex)
            # ì²˜ìŒì—ëŠ” _ _ _ _ _ _ _ _ ë¡œ í‘œì‹œë©ë‹ˆë‹¤.
            #
            # aë¥¼ ì¶”ì¸¡í–ˆë‹¤ë©´ a _ _ _ _ _ _ _ ë¡œ í‘œì‹œë©ë‹ˆë‹¤.
            print(self.display())

            # í˜„ì¬ê¹Œì§€ ì¶”ì¸¡í•œ ê¸€ìë“¤ì„ í‘œì‹œí•©ë‹ˆë‹¤.
            guess = input("ê¸€ìë¥¼ ì¶”ì¸¡í•´ë³´ì„¸ìš”: ").lower()

            # ì¶”ì¸¡í•œ ê¸€ìê°€ ì´ë¯¸ ì¶”ì¸¡í•œ ê¸€ì ëª©ë¡ì— ìˆë‹¤ë©´
            if guess in self.guesses:
                print("ì´ë¯¸ ì¶”ì¸¡í•œ ê¸€ìì…ë‹ˆë‹¤.")

            # ì¶”ì¸¡í•œ ê¸€ìê°€ ì„¤ì •í•œ ë‹¨ì–´ì— í¬í•¨ë˜ì–´ ìˆì§€ ì•Šë‹¤ë©´
            elif guess in self.word:
                # ì¶”ì¸¡í•œ ê¸€ìë¥¼ ì¶”ì¸¡í•œ ê¸€ì ëª©ë¡ì— ì¶”ê°€í•˜ê³  "ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨êµ°ìš”. ë” í˜ë‚´ë´ìš”"ë¼ê³  ì¶œë ¥í•´ì£¼ì„¸ìš”
                self.guesses.add(guess)
                # ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨êµ°ìš”. ë” í˜ë‚´ë´ìš”"ë¼ê³  ì¶œë ¥í•´ì£¼ì„¸ìš”
                # ë§Œì•½ ì¶”ì¸¡í•œ ê¸€ìë“¤ì´ ì„¤ì •í•œ ë‹¨ì–´ì˜ ëª¨ë“  ê¸€ìë¥¼ í¬í•¨í•œë‹¤ë©´
                if set(self.word).issubset(self.guesses):
                    print(f"ì¶•í•˜í•©ë‹ˆë‹¤! ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨ìŠµë‹ˆë‹¤: {self.word}")
                    break
            # ì¶”ì¸¡í•œ ê¸€ìê°€ ì„¤ì •í•œ ë‹¨ì–´ì— í¬í•¨ë˜ì–´ ìˆì§€ ì•Šë‹¤ë©´
            else:
                # ì‹œë„ íšŸìˆ˜ë¥¼ 1 ê°ì†Œì‹œí‚µë‹ˆë‹¤.
                self.attempts -= 1
                print(f"í‹€ë ¸ìŠµë‹ˆë‹¤. ë‚¨ì€ ì‹œë„ íšŸìˆ˜: {self.attempts}")
            # ì‹œë„ íšŸìˆ˜ê°€ 0ì´ ë˜ì—ˆë‹¤ë©´
            if self.attempts == 0:
                print(hangman_pics[-1])
                print(f"ê²Œì„ ì˜¤ë²„. ì •ë‹µì€: {self.word}")


if __name__ == "__main__":
    game = HangmanGame()
    game.play()
```
